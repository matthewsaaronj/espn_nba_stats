```python
html = urlopen('http://www.espn.com/nba/player/gamelog/_/id/3908809/donovan-mitchell')
bsObj = BeautifulSoup(html)
```

```python
bsObj
```

```python
bsObj.find_all('table')
```

```python
bsObj.find_all('table')[1].findAll('td', title=re.compile('[A-Z]'))
```

```python
stat_headers = []
for attr in bsObj.find_all('table')[1].find_all('td', title=re.compile('[A-Z]')):
    if not attr.get_text() in stat_headers:
        stat_headers.append(attr.get_text())
        
stat_headers
```

```python
stat_headers_reorg = []
r = re.compile('[A-Z0-9]+[-]+[A-Z0-9]')

for x in stat_headers:
    if r.match(x):
        m_a = x.split('-')
        stat_headers_reorg.append(m_a[0])
        stat_headers_reorg.append(m_a[1])
    else:
        stat_headers_reorg.append(x)
        
stat_headers_reorg
```
